## train_test_split
Quand on fait du ML, on divise toujours le dataset en 2 parties:
 - un "train set" dont les données sont utilisées pour **entainer le modèle**
 - un "test set" réservé uniquement à **l'évaluation du modèle**
 
En général, on met 80% de données dans le train set et 20% dans le test set.
- on aura alors les tableaux X_train et Y_train qui seront utilisés dans la méthode fit() pour entrainer le modèle.
- mais aussi X_test et Y_test utilisés dans la méthode score() pour évaluer notre modèle

Pour faire cela dans Python, on utilisera la fonction train_test_split() du module model_selection de sklearn.

Pour l'exemple, on va utiliser le dataset des fleurs d'iris:

```python
## on importe d'abord des librairies l'environnement
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
#chargement de base de données wine
iris = load_iris()

X = iris.data
y = iris.target

print(X.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.8)
print(y)
```

Il contient 178 échantillons. Pour diviser ce dataset, on importe la fonction train_test_plit de sklearn

```python
from sklearn.model_selection import train_test_split

# test_size = le % de données test. Ici 0.2 signifie 20%
# X = données
# y = cibles
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(X_train.shape)
print(X_test.shape)
```

On a bien 36 échantillons pour les tests => 36/178 = 20%

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.scatter(X_train[: ,0], X_train[: ,1], c=Y_train, alpha=0.8)
plt.title('Train Set')
plt.subplot(122)
plt.scatter(X_test[: ,0], X_test[: ,1], c=Y_test, alpha=0.8)
plt.title('Test Set')
```

Quand on utilise la fonction train_set_split, le dataset est mélangé de façon aléatoire avant de diviser en 2 parties. On peut contrôler l'aléatoire en fixer l'argument ramdon_state sur un nombre. random_state définit simplement une graine sur le générateur aléatoire, de sorte que vos divisions de test de train soient toujours déterministes. Si vous ne définissez pas de graine, c'est différent à chaque fois.

<p><b>On va maintenant entrainer un modèle de Machine Learning avec KNeighborsClassifier de sklearn</b></p>
<p>La bibliothèque Scikit-learn de Python destinée à l’apprentissage automatique approvisionne le module sklearn.neighbors qui contient les méthodes d’apprentissage basées sur les voisins.

Comme spécifié précédemment, l’algorithme KNN est utilisé ainsi pour la classification plutôt que pour la régression. Ce que nous allons détailler dans ce qui suit.</p>

```python
# import de la librairie
from sklearn.neighbors import KNeighborsClassifier
```

```python
# création d'un modèle avec un voisin de 15 (test)
model = KNeighborsClassifier(n_neighbors=15)

# entrainement du modèle
model.fit(X_train, Y_train)
print('Train score : ', model.score(X_train, Y_train))
print('Test score : ', model.score(X_test, Y_test))
```

## cross_validation

<p>Pour évaluer un modèle, il faut le soumettre à des données qui n'a jamais vues. Or, si on règle nos modèles sur les données du test set, alors ils auront vu indectement ces données puisqu'ils sont réglées dessus. 
Pour cette raison, on découpe une 3ème section dans notre dataset qu'on appellera la <b>validation set</b>. Cette section nous permet de chercher les réglages du modèle qui donnent les meilleurs performances tout en gardant à côté les données du test set pour évaluer la machine sur des données qu'elle n'aura jamais vues.</p>

<p>Donc, quand on veut comparer 2 modèles de machine learning <i>(par exemple avec 2 voisins et un autre avec 7 voisins)</i>, on commencer à entrainer ces modèles sur le train set puis on selectionnera celui qui aura la meilleur performance sur le validation set. Ensuite, on pourra évaluer ce modèle sur le test set afin d'avoir une idées de sa performance dans sa vrai vie.</p>

<p>La  <b>cross validation</b> consiste à entrainer puis valider notre modèle sur plusieurs découpes possible du train set. Par exemple, si on découpe notre train set en 5 partie, on peut entrainer notre modèle sur les 4 premières parties puis le valider sur la 5ème partie. Ensuite, on va cela pour toutes les configurations possible. Au final, on fera la moyenne des 5 scores que l'on obtient. Ainsi, si on veut comparer 2 modèles, alors on sera sûr de prendre celui qui a en moyenne les meilleurs performances.</p>

<p><b>Place au code:</b></p>

<font color=green>Je pense qu'il faut relire un peu tu as des fautes de frappe.Sinon c'est agréable à lire et on te suis bien.</font>

```python
# import de fonction pour découper les tain set
from sklearn.model_selection import cross_val_score

# cv = le nombre de split, scoring = la précision que nous cherchons
cross = cross_val_score(KNeighborsClassifier(), X_train, Y_train, cv=5, scoring='accuracy')

print(cross)
# récupération de la moyenne
print('Moyenne : ', cross.mean())
```

<font color=green>Il peut être utile d'expliquer le tableau "cross" juste au dessus le la moyenne.<font>

On va maintenant afficher les résultats sur un graphique grâce à la méthode validation_curve pour trouver le meilleur voisin pour une meilleur performance.

```python
from sklearn.model_selection import validation_curve

model = KNeighborsClassifier()
k = np.arange(1, 50)

train_score, val_score = validation_curve(model, X_train, Y_train, param_name='n_neighbors', param_range=k, cv=5)

val_score.shape
```

<font color=green>Pareil je pense qu'il peut être bien d'expliquer ce qui est affiché<font>

```python
# création du graphique
plt.plot(k, val_score.mean(axis=1), label='validation')
plt.plot(k, train_score.mean(axis=1), label='train')

plt.ylabel('score')
plt.xlabel('voisins')
plt.legend()
```

On peut remarquer sur le graphique que pour avoir un résultat au dessus de 97%, il faudra un voisin de 11 ou de 12 (0.975 exactement).
